## HTTP 요청 실행

```python
# 관련 라이브러리를 호출합니다.
import requests
```

```python
# 요청 URL을 설정합니다.
url = 'https://finance.naver.com/marketindex/'
```

```python
# HTTP 요청을 실행합니다.
res = requests.get(url = url)
```

## HTTP 응답 확인

```python
# HTTP 응답 상태코드를 확인합니다.
res.status_code
```

```python
# HTTP 응답 헤더를 확인합니다.
res.headers
```

```python
# HTTP 요청 URL만 확인합니다.
res.url
```

```python
# HTTP 응답 Body를 문자열(str)로 출력합니다.
res.text
```

## 필요한 데이터 수집

### HTML Parsing

```python
# 관련 라이브러리를 호출합니다.
from bs4 import BeautifulSoup as bts
```

```python
# str 자료형을 bs4.BeautifulSoup 자료형으로 변환합니다.
soup = bts(markup = res.text, features = 'html.parser')
```

```python
# soup을 출력합니다.
soup
```
