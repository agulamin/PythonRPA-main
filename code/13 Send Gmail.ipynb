## 첨부 파일 추가 발송

```python
# 관련 라이브러리를 호출합니다.
import os
from email.mime.base import MIMEBase
from email import encoders
```

```python
# 현재 작업경로를 확인합니다.
os.getcwd()
```

```python
# data 폴더로 작업경로를 변경합니다.
os.chdir(path = '../data')
```

```python
# 작업경로에 포함된 폴더명과 파일명을 출력합니다.
os.listdir()
```

```python
# 첨부 파일명을 설정합니다.
File = 'Market_Index_Form.xlsx'
```

```python
# 첨부 파일을 추가합니다.
add = MIMEBase(_maintype = 'application', _subtype = 'octet-stream')
add.set_payload(payload = open(file = File, mode = 'rb').read())
encoders.encode_base64(msg = add)
add.add_header(
    _name = 'content-disposition', 
    _value = 'attachment', 
    filename = os.path.basename(p = File)
)
```

```python
# 발송 메시지에 본문을 추가합니다.
msg.attach(payload = add)
```

```python
# 이메일을 발송합니다.
server.sendmail(
    from_addr = User, 
    to_addrs = To.split(','), 
    msg = msg.as_string()
)
```

```python
# SMTP 서버 연결을 끊습니다.
server.close()
```
