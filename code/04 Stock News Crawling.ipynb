- 링크 : https://finance.naver.com/news/mainnews.nhn?date=2021-04-11

## HTTP 요청 실행

```python
# 관련 라이브러리를 호출합니다.
import requests
```

```python
# 요청 URL을 설정합니다.
url = 'https://finance.naver.com/news/mainnews.nhn'
```

```python
# 오늘 날짜를 yyyy-mm-dd 형태의 문자열로 생성합니다.
today = '2021-04-12'
```

```python
# Query String을 딕셔너리로 생성합니다.
query = {'date': today}
```

```python
# HTTP 요청을 실행합니다.
res = requests.get(url = url, params = query)
```

## HTTP 응답 확인

```python
# HTTP 응답 상태코드를 확인합니다.
res.status_code
```

```python
# HTTP 응답 헤더를 확인합니다.
res.headers
```

```python
# HTTP 요청 URL만 확인합니다.
res.url
```

```python
# HTTP 응답 Body를 문자열(str)로 출력합니다.
res.text
```

## 필요한 데이터 수집

### HTML Parsing

```python
# 관련 라이브러리를 호출합니다.
from bs4 import BeautifulSoup as bts
```

```python
# str 자료형을 bs4.BeautifulSoup 자료형으로 변환합니다.
soup = bts(markup = res.text, features = 'html.parser')
```

```python
# soup을 출력합니다.
soup
```

### HTML의 속성값을 리스트로 저장하는 사용자 정의 함수 생성

```python
# 뉴스 제목을 포함하는 HTML 요소만 선택하고 items에 할당합니다.
items = soup.select(selector = 'ul.newsList > li')
```

```python
# items의 길이를 확인합니다.
len(items)
```

```python
# items의 첫 번째 원소만 출력합니다.
items[0]
```

```python
# 뉴스 제목을 포함하는 HTML 요소를 찾아 css에 할당합니다.
# [참고] 뉴스 썸네일이 있으면 'dd', 없으면 'dt' 태그가 사용되므로 태그를 삭제합니다.
css = '.articleSubject > a'
```

```python
# items의 첫 번째 원소에서 뉴스 링크를 선택합니다.
items[0].select(selector = css)
```

```python
# 위 코드를 실행한 결과에서 첫 번째 원소만 선택하고 'href' 파라미터에 지정된 값을 반환합니다.
# [참고] Tag.get_text()로 텍스트를 반환합니다.
items[0].select(selector = css)[0].get(key = 'href')
```

```python
# 리스트 컴프리헨션을 이용하여 items의 모든 원소에 포함된 뉴스 링크를 리스트로 반환합니다.
[item.select(selector = css)[0].get(key = 'href') for item in items]
```

```python
# 리스트 컴프리헨션을 사용하여 HTML 요소의 속성값을 반환하는 사용자 정의 함수를 생성합니다.
# [참고] 사용자 정의 함수를 'getHtmlData.py' 파일로 저장하여 필요할 때마다 호출합니다.
def getAttr(x, css, key):
    return [i.select(selector = css)[0].get(key = key) for i in x]
```
